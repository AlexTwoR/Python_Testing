```python
import numpy as np
import pandas.io.data as web
import datetime as dtime
import pandas as pd
from pandas import DataFrame, Series

%matplotlib inline
```

```python
start=dtime.datetime(2015,1,1)
end=dtime.datetime(2016,1,1)
```

```python
stock1=web.DataReader('SP',"yahoo",start,end)["Adj Close"]
stock1.plot()
```

```python
stock2=web.DataReader('DJX',"yahoo",start,end)["Adj Close"]
stock2.plot()
```

```python
pd.Series.corr(stock1,stock2)
```

```python
stocks=DataFrame({"SnP":stock1, "DJ":stock2})
stocks
stocks2=stocks.copy()
```

```python
stocks2=stocks2.dropna()
```

```python
stocks2['Diff']=(stocks2['SnP']-stocks2['DJ'])
```

```python
stocks2.resample('M').mean()
```

```python
stocks2["DJ"].resample('M').mean()
```
